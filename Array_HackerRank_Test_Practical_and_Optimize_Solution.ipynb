**Problem Assesment**: Given an array, return index from which left and right sum of this number are both equal.

**Example A**: [ 1, 2, 3, 3 ], the answer is 3 because 1+2=3 equals to the last number 3.

**Example B**: [2, 1, 2, 1, 3, 2] , the answer is 4 because 2+1+2=5 equals to the last numbers 3+2=5.


```
#@title Array a is defined as benchmark for this analysis
a = [ 2, 2, 2, 2, 2, 2, 2, 6, 10, 3, 1, 5, 2, 5, 3, 1, 3, 14]
```

---
**function_a**: This function was my original solution to the problem. As we can see, this is a practical approach like a o(n) algorithm because its a loop

```
def function_a(array_):
  n = len(array_)                       #Array size
  m = 0
  for i in range(n):                    #Iterate from first to last element in array
    sum_left = sum(array_[0:i+1])       #Increase left side array in order to increase sum
    sum_right = sum(array_[i:n+1])      #Decrease right side array to make a smaller result
    if sum_left == sum_right: m=i       #Compare both size if it equals point m as the right index, otherwise continue on loop
  
  return m
```

```
function_a(a)                          #Results for array a from function a
```

**function_b**: This last and optimize function takes inspiration from a tree perspective where sums began on the arrayÂ´s middle and recursively goes down on takind mid-sizes arrays o(n^2).

![alt text](https://pythonschool.net/data-structures-algorithms/images/tree.jpg)

In this tree, Each point represents a leaf, if the leaf founds sum_left and sum_right are equal stops cicle. Otherwise, continue on left or right leaf depending if its bigger or smaller the mid-size array than the other one (look at code).



```
def leaf(start, end, sum_left, sum_right, i, array_):

  if sum_left > sum_right:                                    #If left array sum is bigger than right continue on left leaf
    end = i                                                   #Last element points to i index in order to ignore other halfway numbers
    i = start + round((end - start)/2) - 1                    #Create new start and end subarray for nex cicle
    if end == i : i = end - 1
    sum_left  -= sum(array_[i:end])                           #Decrease sum_left
    sum_right += sum(array_[i+1:end+1])                       #Increase sum_right to eliminate gap to sum_left
    return leaf(start, end, sum_left, sum_right, i, array_)   #Continue to left leaf with new parameters

  if sum_left < sum_right:                                    #If right array sum is bigger than left continue on right leaf
    start = i                                                 #Start element points to i index in order to ignore other halfway numbers
    i = start + round((end - start)/2) - 1                    #Create new start and end subarray for nex cicle
    if start == i : i = start + 1
    sum_left  += sum(array_[start:i])                         #Increase sum_left
    sum_right -= sum(array_[start+1:i+1])                     #Decrease sum_right to eliminate gap to sum_left
    return leaf(start, end, sum_left, sum_right, i, array_)   #Continue to right leaf with new parameters

  return i                                                    #Break cicle if sum_left and sum_right are equal
```

```
def function_b(array_):                                       #Initialize function_b by middle array left and right sums
  start, end = 0, len(array_)
  i = start + round((end - start)/2) - 1
  sum_left= sum(array_[start:i])
  sum_right= sum(array_[i+1:end])
  
  return leaf(start, end, sum_left, sum_right, i, array_)     #Start leaf cicle
```
